EXPLORATRY DATA ANALYSIS

```python
df.columns #printout all columns for refernce
```

Region on the basis of employed people

```python
sns.set(rc={'figure.figsize':(20,5)})
mani=df.groupby(['Region'],as_index=False)[' Estimated Employed'].sum().sort_values(by=' Estimated Employed',ascending=False).head(15)
sns.barplot(data=mani,x='Region',y=' Estimated Employed')
plt.title("ESTIMATED EMPLOYEES OVER STATE")
plt.xlabel("States ")
```

Here we can see that uttar Pradesh has highest number of employed people

Region on the basis of Labour participation rate

```python
sns.set(rc={'figure.figsize':(15,5)})
mani_lab=df.groupby(['Region'],as_index=False)[' Estimated Labour Participation Rate (%)'].sum().sort_values(by=' Estimated Labour Participation Rate (%)',ascending=False).head(10)
sns.barplot(data=mani_lab,x='Region',y=' Estimated Labour Participation Rate (%)')
plt.title("LABOUR PARTICIPATION RATE OVER STATES")
plt.xlabel("state")
```

```python
sns.set(rc={'figure.figsize':(15,5)})
mani_unemp=df.groupby(['Region'],as_index=False)[' Estimated Unemployment Rate (%)'].sum().sort_values(by=' Estimated Unemployment Rate (%)',ascending=False).head(10)
sns.barplot(data=mani_unemp,x='Region',y=' Estimated Unemployment Rate (%)')
plt.title("UNEMPLOYMENT RATE VS STATE")
```

Here we can see that 'Haryana' State has highest Unemployement Rate (%)

```python

mani_date=df.groupby([' Date'],as_index=False)[' Estimated Unemployment Rate (%)'].sum().head(15)
sns.lineplot(data=mani_date,x=' Date',y=' Estimated Unemployment Rate (%)', )
plt.title('UNEMPLOYMENT RATE OVER MONTHS IN YEAR 2020')
```

Here we can see in month 'MAY' there is huge increase in Unemployment rate. 

```python
sns.set(rc={'figure.figsize':(15,5)})
mani_reg1=df.groupby(['Region.1'],as_index=False)[' Estimated Unemployment Rate (%)'].sum().sort_values(by=' Estimated Unemployment Rate (%)',ascending=False).head(15)
sns.barplot(data=mani_reg1,x='Region.1',y=' Estimated Unemployment Rate (%)')
plt.title("UNEMPLOYMENT RATE OVER REGION 1")
```

```python
sns.boxplot(data=mani_unemp,x='Region',y=' Estimated Unemployment Rate (%)')
plt.title("UNEMPLOYMENT RATE VS STATE")
plt.xlabel("STATE")
```

Overall the 'Haryana' State has high rate of UNemployement
